```python
import requests

def search_contributors(location):
    url = "https://api.github.com/search/users"
    params = {
        "q": f"location:{location}",
        "per_page": 100,
        "page": 1
    }
    headers = {
        "Accept": "application/vnd.github.v3+json"
    }
    contributors = []
    while True:
        response = requests.get(url, params=params, headers=headers)
        if response.status_code == 200:
            data = response.json()
            contributors.extend(data["items"])
            if "next" in response.links:
                url = response.links["next"]["url"]
                params = {}
            else:
                break
        else:
            print(f"Error: {response.status_code}")
            break
    return contributors

location = "Egypt"
contributors = search_contributors(location)
print(len(contributors))
contributors
for contributor in contributors:
    print(contributor["login"])
```

```python
# load frictionless datapackage from github
from frictionless import Package
import petl as etl

package = Package("https://raw.githubusercontent.com/github/innovationgraph/main/datapackage.json")

# print package and schema
print(package)
```

```python
 
developers = package.get_resource("developers").read_cells()
# filter EG and sort by year and quarter
eg_devs = etl.select(developers, "{iso2_code} == 'EG'").sort(key=('year', 'quarter'))
eg_devs.displayall()
```

```python
 
languages = package.get_resource("languages").read_cells()
# filter EG and sort by year and quarter
eg_devs = etl.select(languages, "{iso2_code} == 'EG'").sort(key=('year', 'quarter'))
eg_devs.display()
```
